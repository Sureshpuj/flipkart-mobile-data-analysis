```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sys
```

```python
dt=pd.read_csv("Flipcart_Mobile.csv")
dt
```

```python
print("rows and columns :",dt.shape)
```

```python
dt.info()
```

```python
cdt=dt.fillna(0)
cdt
```

```python
cdt.isnull().sum()
```

```python
cdt['Discount'] = cdt['Discount'].astype(str).str.extract('(\d+)').astype(float)
cdt
```

```python
cdt['Rating Count'] = cdt['Rating Count'].astype(str).str.extract('(\d+)').astype(float)
cdt
```

```python
cdt['Review Count'] = cdt['Review Count'].astype(str).str.extract('(\d+)').astype(float)
cdt
```

```python
cdt['Battery Capacity'] = cdt['Battery Capacity'].astype(str).str.extract('(\d+)').astype(float)
cdt['Battery Capacity']
```

```python
cdt['Processor'] = cdt['Processor'].astype(str).str.extract('(\d+)').astype(float)
cdt['Processor']
```

```python
cdt.describe()
```

```python
print("Total duplicate: ",cdt.duplicated().sum())
```

```python
plt.title("Molile  Price by Discount(% Off) ",color="#333333")
x=cdt["Discount"].head(10)
y=cdt["Price"].head(10)
plt.bar(x,y, color="#8897AA",edgecolor="black",linewidth=1.5)
plt.ylabel("Price in RS.")
plt.xlabel("Discount % off")
# plt.show()
plt.gca().set_facecolor("#D7E2EB")
plt.show()
```

```python
plt.title("Processor by Price")

x = cdt["Processor"] 
# Discounts as data for the histogram
plt.hist(x,color="#8897AA", edgecolor="black", linewidth=1) 

plt.ylabel("processor")  
plt.xlabel("Price")

plt.gca().set_facecolor("#D7E2EB")  # Set background color
plt.show()
```

```python
cdt["Battery Capacity"] = cdt["Battery Capacity"].astype(str)  # Convert all values to string
cdt["Battery Capacity"] = cdt["Battery Capacity"].str.extract('(\d+)')  # Extract numeric part
cdt["Battery Capacity"] = cdt["Battery Capacity"].fillna(0).astype(int)  # Handle NaN and convert to int
cdt["Battery Capacity"]
```

```python
cdt.info()
```

```python

plt.scatter(cdt["Battery Capacity"], cdt["Avg Rating"], color="#8897AA", edgecolor="black", linewidth=0.9) 

plt.xlabel("Battery Capacity in mAh Li-ion Battery")
plt.ylabel("Average Rating")
plt.title("Battery Capacity vs Average Rating")
plt.gca().set_facecolor("#D7E2EB")
plt.grid(True)
plt.show()
```

```python
filtered_data = cdt[cdt["Price"] > 50000]
# Sorting by Price in descending order and selecting top 10
top_10 = filtered_data.sort_values(by="Price").head(10)
x = top_10["Model Name"] 
y = top_10["Price"]  
plt.bar(x, y, color="#8897AA", edgecolor="black")
plt.title("Top 10 Phone Models Above â‚¹50000 by Price", color="#333333")
plt.xticks(rotation=45, ha="right", fontsize=5)
plt.xlabel("Mobile Model", fontsize=12)
plt.gca().set_facecolor("#D7E2EB")
plt.ylabel("Price (in â‚¹)")
plt.show()
```

```python
plt.title("Spread of Avg Rating Across Different Phone Models",fontweight="bold")
x = cdt["Model Name"].head()
y = cdt["Avg Rating"]
sys.boxplot(x=x, y=y, palette="pastel")  
plt.xlabel("Phone Model", fontsize=12)
plt.ylabel("Average Rating", fontsize=12)
plt.gca().set_facecolor("#D7E2EB")
plt.xticks(rotation=45, ha="right",fontsize=5)
plt.show()
```

```python
corr_matrix = cdt[["Price", "Avg Rating", "Battery Capacity"]].corr()

sys.heatmap(corr_matrix, cmap="coolwarm", fmt=".2f", linewidths=10)

plt.title("Correlation Matrix: Price, Ratings, and Battery",fontweight="bold")
plt.show()
```

```python

price = cdt["Price"].head(10)
processor_counts = cdt["Processor"]  # Extracting processor names

plt.pie(price, labels=processor_counts.head(10), autopct="%1.1f%%", 
        colors=["#ff9999", "#66b3ff", "#99ff99", "#ffcc99", "#c2c2f0"], 
        startangle=140, shadow=True, wedgeprops={"edgecolor": "black"})

plt.title("Price Distribution of Top 10 Phones by Processor", fontsize=14, fontweight="bold", color="#333333")
plt.show()
```

```python
# plt.title("")
sys.pairplot(cdt,hue='Memory Details',palette='bright')
# plt.gca().set_facecolor("#D7E2EB")
```

```python
# Select top 10 rows based on 'Battery Capacity'
top_10 = cdt.nlargest(10, 'Battery Capacity')
# Create a boxplot using the filtered data
# plt.plot(x='Battery Capacity', y='Lens Detail', data=top_10, palette='Blues')
plt.bar(top_10["Battery Capacity"], top_10["Lens Detail"], color='black')
plt.xlabel("Battery Capacity")
plt.ylabel("Lens Detail")
plt.title("Battery Capacity vs Lens Detail (Top 10)")
plt.yticks(rotation=45, ha="right",fontsize=5)
plt.gca().set_facecolor("#D7E2EB")
```

```python
x = cdt["Battery Capacity"].head(10) 
y = cdt["Discount"].head(10)
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Discount (%)")
plt.title("Battery Capacity vs Discount")

plt.plot(x, y, marker='o', linestyle='-.', color='#333333', label="Discount Trend",markerfacecolor='red')
plt.gca().set_facecolor("#D7E2EB")
```
